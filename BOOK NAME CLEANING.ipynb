NAMES CORRECTION FOR SEARCH OPTIMISATION

```python
lsrep=[]
for i in df['Name of Book']:
    rep=''
    if ':' in i and ',' in i:
        ls=i.split(':')
        ls2=ls[0].split(',')
        ls2=ls2[::-1]

        
        for j in ls2:
            rep=rep+j+' '
        rep=rep+':'+' '+ls[1]
        
        
    elif ',' in i and ':' not in i:
        ls=i.split(',')
        ls=ls[::-1]
        
        for k in ls:
            rep=rep+' '+k
    else:
        rep=i

    lsrep.append(rep)
```

```python
final_ls=[]
for j in lsrep:
    j=j.strip()
    j=j.lower()
    final_ls.append(j)
    
final_ls
```

```python
df['Name of Book']=final_ls
```
